```python
iterations, paths, bf, temps, prob = simulated_annleaing(T=50, points=5, dim=10, T_bound=0.002, T_decay=0.999, max_iter=10000, T_schedule='exp', seed=0)
```

```python
plt.scatter(*paths[0], color='r', zorder=1000)
plt.plot(*paths[1], color='b', zorder=50)
plt.title('Current Solution')
plt.savefig('xc.png')
```

```python
plt.scatter(*paths[0], color='r', zorder=1000)
plt.plot(*paths[1], color='b', zorder=50)
plt.plot(*paths[0], color='g', zorder=1, linestyle='dotted')
plt.title('Permutation')
plt.savefig('permute.png')
```

```python
for i in range(0, len(iterations), 2):
    plt.subplot('221')
    plt.title(f'Path at Temperature {temps[i]:.4f}')
    plt.scatter(*paths[i], color='r', zorder=1000)
    plt.plot(*paths[i], color='b')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    plt.subplot('222')
    plt.plot(bf[:i], color='r')
    plt.title('Total Distance')
    plt.xlabel('Iteration')
    plt.ylabel('Distance')
    
    plt.subplot('223')
    plt.scatter(iterations[:i], prob[:i], color='g', marker='*')
    plt.title('Acceptance Probability')
    plt.ylim(-0.2, 1.2)
    plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.xlabel('Iteration')
    plt.ylabel('Probability')
    
    plt.subplot('224')
    plt.title('Temperature Schedule')
    plt.plot(iterations[:i], temps[:i], color='orange')
    plt.xlabel('Iteration')
    plt.ylabel('Temperature')
    
    plt.pause(1e-9)
    plt.savefig(f'out/{int(i/2):05d}.png')
    plt.show()
    plt.close()
```
