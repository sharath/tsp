```python
from generate import simulated_annleaing
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
f = lambda x : x*np.sin(x) + 100

x = np.linspace(0, 4*np.pi, 1000)
y = f(x)

plt.plot(x, y, color='black')
plt.scatter([3], [f(3)], label='Starting Position', color='y', zorder=100)
plt.scatter([4.81447], [f(4.81447)], label='Local Minimum', color='r', zorder=100)
plt.scatter([11.0407], [f(11.0407)], label='Global Minimum', color='teal', zorder=100)

plt.xticks([]), plt.yticks([])
plt.legend()
plt.box(False)
plt.savefig('hillclimbing.png', dpi=150)
plt.show()
```

```python
iterations, paths, bf, temps, prob = simulated_annleaing(T=100, points=16, dim=1, T_bound=0, T_decay=0.999, max_iter=10000, T_schedule='exp', seed=10)
```

```python
plt.subplot('121')
k = paths[0]
k[0].append(k[0][0]), k[1].append(k[1][0])
plt.scatter(*k, color='black', zorder=1000)
plt.plot(*k, color=(0.3,0.3,0.3))
plt.xticks([]), plt.yticks([])
plt.xlabel('Starting Cycle')
plt.box(False)

plt.subplot('122')
k = paths[-1]
k[0].append(k[0][0]), k[1].append(k[1][0])
plt.scatter(*k, color='black', zorder=1000)
plt.plot(*k, color=(0.3,0.3,0.3))
plt.xlabel('Final Cycle')

plt.xticks([]), plt.yticks([])
plt.box(False)

plt.savefig('sasolution.png', dpi=150)
plt.show()
```

```python
plt.plot(bf, color=(0.4,0.4,0.4))
plt.xticks([]), plt.yticks([])
plt.ylabel('Distance', rotation=0, labelpad=20)
plt.xlabel('Time')
plt.box(False)
plt.savefig('distances.png', dpi=150)
plt.show()
```

```python
plt.scatter(iterations, prob, color=(0.4,0.4,0.4), marker='.')
plt.xticks([])
plt.ylabel('Probability', rotation=0, labelpad=30)
plt.xlabel('Time')
plt.box(False)
plt.savefig('probs.png', dpi=150)
plt.show()
```
