```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x : x*np.sin(x)
```

```python
x = np.linspace(0, 4*np.pi, 1000)
y = f(x)
```

```python
plt.plot(x, y, color='b')
plt.scatter([3], [f(3)], label='Starting Position', color='y', zorder=100)
plt.scatter([4.81447], [f(4.81447)], label='Local Minimum', color='r', zorder=100)
plt.scatter([11.0407], [f(11.0407)], label='Global Minimum', color='g', zorder=100)
plt.title('Sensitivity to Starting Position')
plt.legend()
plt.savefig('hillclimbing.png', dpi=100)
plt.show()
```
